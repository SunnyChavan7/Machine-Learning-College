```python
# to create new column
sex = pd.get_dummies(df["Sex"], drop_first=True)
```

```python
sex.head()
```

```python
df.head()
```

```python
embark = pd.get_dummies(df["Embarked"], drop_first=True)
```

```python
embark.head()
# Q S value
# 0 0  C 
  1  
```

```python
pcl = pd.get_dummies(df["Pclass"], drop_first=True)
```

```python
df = pd.concat([df, sex, embark, pcl], axis=1)
```

```python
df.head()
```

```python
# remove the earlier columns and unnecessary columns
df.drop(["PassengerId",  "Pclass", "Name", "Sex", "Ticket", "Embarked"], axis=1, inplace=True)
```

```python
df.head()
```

```python
# y : output   x : input
y= df["Survived"]
x= df.drop(["Survived"], axis=1)
```

```python
df.head()
```

```python
# Train-test split the data (x and y)
from sklearn.model_selection import train_test_split
x.columns = x.columns.astype(str)   # as column names must be string values only
                                    # here we have 2 and 3 as integers 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)
```

```python
# importing model
from sklearn.linear_model import LogisticRegression
```

```python
# creating instance of the model
logreg = LogisticRegression()
```

```python
# fitting the model to training data
logreg.fit(x_train, y_train)
```

```python
# passing the test data and getting predicted values from the model
prediction = logreg.predict(x_test)
```

```python
# To get accuracy of the predicted data
from sklearn.metrics import accuracy_score
```

```python
# final accuracy score
accuracy_score(y_test, prediction)
```
